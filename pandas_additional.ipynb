```python
import pandas as pd
```

```python
data = pd.read_csv("/home/olga/Documents/teaching/machine_Learning/kaggle/hause_pricing/data/train.csv", index_col=0)
```

```python
data.head()
```

```python
df = pd.DataFrame(columns = ['YrSold', 'SalePrice', 'LotArea']) 
for i, year in enumerate(data.YrSold.unique()):
    selector_n = (data.Street == 'Pave') & (data.YrSold == year)
    a = data[selector_n].SalePrice.mean()
    b = data[selector_n].LotArea.mean()
    df.loc[i] = [year, a, b]
```
