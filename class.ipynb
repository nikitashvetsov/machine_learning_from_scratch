 # \_\_new__ vs \_\_init

Use \_\_new__ when you need to control the creation of a new instance. Use \_\_init__ when you need to control initialization of a new instance.

\_\_new__ is the first step of instance creation. It's called first, and is responsible for returning a new instance of your class. In contrast, \_\_init__ doesn't return anything; it's only responsible for initializing the instance after it's been created.

```python
class A:
    
    def __new__(cls):
        print("A.__new__ is called")
        
    def __init__(self):
        self.a = "a"
        print("A.__init__ called")


a = A()
print(a.a)
```

```python
class Car():
        
    def __init__(self, model, year):  
        self.model = model
        self.yeat = year
        
car = Car("mersedes", 2015)
print(car.model)
```

как сделать несколько конструкторов? никак

или делаем много if в init или делаем фабрику. заодно что такое метод класса рассмотрим

```python
class Car():
        
    def __init__(self):  
        self.model = ""
        self.yeat = -1
        
    @classmethod
    def create_car(cls, from_what, argv):
        if (from_what == "other_car"):
            ## check argv
            car = cls.create_from_car(argv[0])
            return car
        if (from_what == "data"):
            ## check argv
            car = cls.create_from_data(argv[0], argv[1])
            return car
            
    @classmethod
    def create_from_car(cls, other_car):
        car = cls()
        car.model = other_car.model
        car.yeat = other_car.year
        return car
    
    @classmethod
    def create_from_data(cls, model, year): 
        car = cls()
        car.model = model
        car.yeat = year
        return car
        
car = Car.create_car("data",["mersedes", 2015])
print(car.model)
```
