 # \_\_new__ vs \_\_init

Use \_\_new__ when you need to control the creation of a new instance. Use \_\_init__ when you need to control initialization of a new instance.

\_\_new__ is the first step of instance creation. It's called first, and is responsible for returning a new instance of your class. In contrast, \_\_init__ doesn't return anything; it's only responsible for initializing the instance after it's been created.

```python
class Car():
        
    def __init__(self):  
        self.model = ""
        self.yeat = -1
        self._velocity = 10
        
    @property
    def velosity(self):
        return self._velocity
    
    @velosity.setter
    def velosity(self, vel):
        if vel >= 0:
            self._velocity = vel
        else:
            print("wrong velocity")
            
    @velosity.deleter
    def velosity(self):
        print("delete velocity")
        del self._velocity
```

```python
car = Car()
print(car.velosity)
car.velosity = 90
print(car.velosity)
del car.velosity
print(car.velosity)
```

```python
def log(func):
    """
    Логируем какая функция вызывается.
    """
    
    def wrap_log(*args, **kwargs):
        name = func.__name__
        print("Вызов функции: %s" % name)
        result = func(*args, **kwargs)
        print("Результат: %s" % result)
        return func
    
    return wrap_log
 
 
@log
def double_function(a):
    """
    Умножаем полученный параметр.
    """
    return a*2
 
 
value = double_function(2)
```
